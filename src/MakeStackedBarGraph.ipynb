```python
#<損益計算書>
#貸方
pl_credit = {
    "pl_売上高": 125517
}

#借方
pl_debit={
     "pl_売上原価": 11795,
     "pl_販管費": 85832, 
     "pl_営業利益": 27888
}

PL= (pl_credit, pl_debit)
```

```python
 #<貸借対照表>
#貸方
bs_credit = {
"bs_現預金等": 178094,
"bs_売上債権": 5971,
"bs_棚卸資産": 51646,
"bs_その他流動資産": 4402,
"bs_有形固定資産": 131082,
"bs_無形固定資産": 1023,
"bs_投資その他": 35672,
}

bs_debit = {
"bs_買掛債務": 18661,
"bs_有利子負債": 0,
"bs_その他流動負債": 18497,
"bs_その他固定負債": 4921,
"bs_純資産":365901,
}

BS= (bs_credit, bs_debit)
```

```python
BS[0]
```

```python
bar_idx = 0
label_li = []

for tmp in [BS, PL]:
    for finanfial_table in tmp:
        cumulative_data = 0
        for item_name, amount in  finanfial_table.items() :
            plt.bar(bar_idx, amount, bottom=cumulative_data,  alpha=0.7)
            cumulative_data += amount
            label_li.append(item_name)
        bar_idx += 1
    
plt.legend(label_li, bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=13)
plt.ylabel("百万円", fontsize=17)
#軸の横幅調整用のダミーplot;
plt.title('ZOZO　2020年3月期決算', fontsize=14) # タイトル
plt.plot([0,2],[0,0],color='#00000000')
plt.show()
```
